## Librer√≠as

```python
import matplotlib.pyplot as plt
import matplotlib.image as pmimg
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import cv2
```

## Dataset

```python
img = cv2.imread('datasets/PlayeraVerde.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15,5))
plt.imshow(img)
plt.show()
```

## RGB --> HSV

```python
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

plt.figure(figsize=(15,5))
plt.imshow(hsv)
plt.show()
print(hsv[1200,600])
```

## Detecci√≥n color verde y cambio a azul

```python
for i in hsv:
    for j in i:
        if (j[0] >= 60 and j[0] <= 90) and (j[1] >= 70 and j[1] <= 255) and (j[2] >= 60 and j[2] <= 255):
            j[0] = j[0] + 30


resultado = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(resultado)
plt.show()
```
