## Librerías

```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
import seaborn as sns

# Datasets
from sklearn.datasets import load_diabetes

# Training and testing sets
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold

# Regression models
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor

# Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
from scipy.stats import pearsonr
```

## Dataset

```python
data = pandas.read_csv('datasets/winequality-red.csv', sep=',')
data
```

## First experiment

### - Randomly divide the data into training (70%) and testing (30%) sets
### - Generate three regression models: 

###### - Linear Regression (from sklearn.linear_model import LinearRegression)
##### - Random Forest (from sklearn.ensemble import RandomForestRegressor)
##### - MultiLayer Perceptron (from sklearn.neural_network import MLPRegressor)




### - For each model print

##### - MAE
##### - MSE
##### - R2 score, coefficient of determination
##### - Boxplot of the quality
##### - Scatter plot of real and predicted outputs

### 1-. Linear Regression 

```python
X = data.loc[:,'fixed acidity':'alcohol']
Y = data.loc[:,'quality']
X = X.values
Y = Y.values

Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.30, random_state=0)

model = LinearRegression()
model.fit(Xtrain, Ytrain)
Ypred = model.predict(Xtest)

# Ypred = np.round_(Ypred, decimals = 0)

print('MAE:', mean_absolute_error(Ytest, Ypred))
print('MSE:', mean_squared_error(Ytest, Ypred))
print('R2 score:', r2_score(Ytest, Ypred))

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.scatter( Ytest, Ypred )

plt.subplot(1,3,2)
plt.boxplot(Ytest)

plt.subplot(1,3,3)
plt.boxplot(Ypred)

plt.show()
```

### 2-. Random Forest 

```python
X = data.loc[:,'fixed acidity':'alcohol']
Y = data.loc[:,'quality']
X = X.values
Y = Y.values

Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.30, random_state=0)

model = RandomForestRegressor(random_state=0)
model.fit(Xtrain, Ytrain)
Ypred = model.predict(Xtest)

# Ypred = np.round_(Ypred, decimals = 0)

print('MAE:', mean_absolute_error(Ytest, Ypred))
print('MSE:', mean_squared_error(Ytest, Ypred))
print('R2 score:', r2_score(Ytest, Ypred))

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.scatter( Ytest, Ypred )

plt.subplot(1,3,2)
plt.boxplot(Ytest)

plt.subplot(1,3,3)
plt.boxplot(Ypred)

plt.show()
```

### 3-. MultiLayer Perceptron

```python
X = data.loc[:,'fixed acidity':'alcohol']
Y = data.loc[:,'quality']
X = X.values
Y = Y.values

Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.30, random_state=0)

model = MLPRegressor(random_state=0)
model.fit(Xtrain, Ytrain)
Ypred = model.predict(Xtest)

# Ypred = np.round_(Ypred, decimals = 0)

print('MAE:', mean_absolute_error(Ytest, Ypred))
print('MSE:', mean_squared_error(Ytest, Ypred))
print('R2 score:', r2_score(Ytest, Ypred))

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.scatter( Ytest, Ypred )

plt.subplot(1,3,2)
plt.boxplot(Ytest)

plt.subplot(1,3,3)
plt.boxplot(Ypred)

plt.show()
```

## Second experiment

### - Randomly divide the data into 5 splits using K-fold
### - For each iteration:

##### - Create three models: Linear Regression, Random Forest, and MultiLayer Perceptron
##### - Calculate and save for each model the R2 score



### - Generate a visualization with three boxplots (each one per type of model) showing the values of R2 score

```python
data = pandas.read_csv('datasets/winequality-red.csv', sep=',')
X = data.loc[:,'fixed acidity':'alcohol']
Y = data.loc[:,'quality']

X = X.values
Y = Y.values

kf = KFold(n_splits=5, random_state=0, shuffle=True)

LRR2 = []
RFR2 = []
MPR2 = []

for train_idx, test_idx in kf.split(X):
#     print(test_idx)
    Xtrain, Ytrain = X[ train_idx,:], Y[train_idx]
    Xtest, Ytest = X[ test_idx,:], Y[test_idx]
    
    # Linear Regression
    model = LinearRegression()
    model.fit(Xtrain, Ytrain)
    Ypred = model.predict(Xtest)
#     Ypred = np.round_(Ypred, decimals = 0)
    
    LRR2.append(r2_score(Ytest, Ypred))
    
    # Random Forest
    model = RandomForestRegressor(random_state=0)
    model.fit(Xtrain, Ytrain)
    Ypred = model.predict(Xtest)
#     Ypred = np.round_(Ypred, decimals = 0)
    
    RFR2.append(r2_score(Ytest, Ypred))

    # MultiLayer Perceptron
    model = MLPRegressor(random_state=0)
    model.fit(Xtrain, Ytrain)
    Ypred = model.predict(Xtest)
#     Ypred = np.round_(Ypred, decimals = 0)
    
    MPR2.append(r2_score(Ytest, Ypred))
    
print('Finish')
```

```python
df = pandas.DataFrame({'Linear Regression': LRR2,
                       'Random Forest': RFR2,
                       'MultiLayer Perceptron': MPR2})
df
```

```python
plt.figure()

plt.figure(figsize=(10, 5))

sns.boxplot(data=df)
```
